```python
# PJT 2: 자료가 많은 경우

import jason
import pprint

musics_file = open('data/musics.json', encoding='utf-8')
musics_list = json.load(musics_file)

def music_info(musics_list):
    result = [] # list 형태
    for music in musics_list:
        info = {} # dict 형태
        singer = music.get('singer')
        title = music.get('title')
        info['singer'] = singer
        info['title'] = title
        
        # result.append(info)
        # result += info 로 넣으면 안 된다. key 값들만 하나씩 들어간다.
        # ['가'] + ['나'] 하나의 요소로 연결을 한다. ['가', '나']
        result += [info]
    return result
    
print(music_info(musics_list))
```
